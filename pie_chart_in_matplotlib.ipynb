```python
import matplotlib.pyplot as plt
labels = ['A','B','C','D']
sizes = [30,20,40,10]
plt.pie(sizes,labels=labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
colors=['lightcoral','gold','lightskyblue','lightgreen']
labels = ['A','B','C','D']
sizes = [30,20,40,10]
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
#add explode
import matplotlib.pyplot as plt
colors=['lightcoral','gold','lightskyblue','lightgreen']
labels = ['A','B','C','D']
sizes = [30,20,40,10]
explode=(0,0,0.1,0)
#if you want select one piece of this chart you shoul put .1 in select part
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=140,explode=explode)
plt.title('Pie Chart')
plt.show()
```
